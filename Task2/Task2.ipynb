```python
import pandas
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_absolute_error, r2_score, mean_absolute_percentage_error
from matplotlib import pyplot
import numpy as np
```

```python
dataset = pandas.read_csv("housing.csv", sep='\\s+')

target = dataset["MEDV"]
```

```python
# 2 признака

X = dataset.iloc[:, 0:2]
regression = LinearRegression()
regression.fit(X, target)

predicted = regression.predict(X)

print("MAE:", mean_absolute_error(target, predicted))
print("R2:", r2_score(target, predicted))
print("MAPE:", mean_absolute_percentage_error(target, predicted))

figure = pyplot.figure()

ax = figure.add_subplot(1, 3, 1, projection="3d")
ax.plot(X.iloc[:, 0], X.iloc[:, 1], predicted)

ax = figure.add_subplot(1, 3, 2, projection="3d")
ax.plot(X.iloc[:, 0], X.iloc[:, 1], target)

X0, X1 = np.meshgrid([X.iloc[:, 0].min(), X.iloc[:, 0].max()], [X.iloc[:, 1].min(), X.iloc[:, 1].max()])
ax = figure.add_subplot(1, 3, 3, projection="3d")
ax.plot_surface(X0, X1, regression.predict(np.hstack((X0.reshape(-1,1), X1.reshape(-1,1)))).reshape(X0.shape))
```

Как можно заметить, построенная модель (первая) отчасти повторяет форму исходной
