```python
from sklearn.datasets import load_iris, make_classification
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
import pandas
from matplotlib import pyplot
import seaborn
```

```python
iris = load_iris()
X = pandas.DataFrame(iris.data, columns=iris.feature_names)
Y = iris.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
regression = LogisticRegression(random_state=0)
regression.fit(X_train, Y_train)

Y_predicted = regression.predict(X_test)
```

```python
fig, (ax1, ax2) = pyplot.subplots(2)
ax1.scatter(X['petal length (cm)'], X['petal width (cm)'], c=Y)
ax1.set_title("Initial data")
ax2.scatter(X_test['petal length (cm)'], X_test['petal width (cm)'], c=Y_predicted)
ax2.set_title("Test-only data")
pyplot.show()
```
